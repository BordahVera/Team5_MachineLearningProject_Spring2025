```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.decomposition import PCA

# === Load training data ===
df = pd.read_csv("synthetic_bacterial_chemotaxis_data.csv")
X = df.drop(columns=["state"])
y = df["state"]
y_encoded = y.map({"run": 0, "tumble": 1})

# === Train the classifier ===
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X, y_encoded)

# === Predict on training data ===
y_pred = clf.predict(X)
predicted_label = pd.Series(y_pred).map({0: 'run', 1: 'tumble'})
actual_label = y_encoded.map({0: 'run', 1: 'tumble'})

# === Evaluate ===
print("Confusion Matrix:\n", confusion_matrix(y_encoded, y_pred))
print("\nClassification Report:\n", classification_report(y_encoded, y_pred))

# === PCA for visualization ===
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# === Build DataFrame ===
plot_df = pd.DataFrame({
    "PC1": X_pca[:, 0],
    "PC2": X_pca[:, 1],
    "Predicted": predicted_label,
    "Actual": actual_label
})
plot_df["Correct"] = plot_df["Predicted"] == plot_df["Actual"]
misclassified = plot_df[~plot_df["Correct"]]

# === Plot ===
plt.figure(figsize=(10, 7))
sns.scatterplot(data=plot_df, x="PC1", y="PC2", hue="Predicted", palette="Set1", s=100)

# Add misclassified marker only if needed
if not misclassified.empty:
    plt.scatter(misclassified["PC1"], misclassified["PC2"],
                facecolors='none', edgecolors='black', s=150, linewidths=2, label="Misclassified")
    legend_title = "Class"
    plt.legend(title=legend_title, bbox_to_anchor=(1.05, 1), loc='upper left')
else:
    plt.legend(title="Class", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title("Random Forest Classification on Training Data (PCA projection)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.tight_layout()
plt.savefig("training_classification_final.png")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# === 5. Predict on new input dataset from CSV ===
new_data = pd.read_csv("new_bacterial_data_50.csv")  # Replace with your actual path

# Predict with trained classifier
predicted_class = clf.predict(new_data)
predicted_label = ['run' if label == 0 else 'tumble' for label in predicted_class]
new_data["Predicted State"] = predicted_label

# === 6. Show tabular results ===
print("\nPredictions for new data:")
print(new_data)

# === 7. PCA projection using training PCA model ===
new_data_pca = pca.transform(new_data.drop(columns=["Predicted State"]))

plot_df_input = pd.DataFrame({
    "PC1": new_data_pca[:, 0],
    "PC2": new_data_pca[:, 1],
    "Predicted": new_data["Predicted State"]
})

# === 8. Plot ===
plt.figure(figsize=(10, 7))
sns.scatterplot(data=plot_df_input, x="PC1", y="PC2", hue="Predicted", palette="Set1", s=100)
plt.title("Predicted Classification of New Input Data (PCA projection)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Predicted State", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("input_data_classification.png")
plt.show()

# === 9. Save predictions ===
new_data.to_csv("predicted_bacterial_states.csv", index=False)
print("\nPredictions saved to 'predicted_bacterial_states.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# === Step 1: Load and train on labeled data ===
df_train = pd.read_csv("synthetic_bacterial_chemotaxis_data.csv")
X_train = df_train.drop(columns=["state"])
y_train = df_train["state"].map({"run": 0, "tumble": 1})

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# === Step 2: Load input data ===
df_input = pd.read_csv("new_bacterial_data_50.csv")
features = X_train.columns
X_input = df_input[features]

# === Step 3: Predict and calculate confidence ===
proba = clf.predict_proba(X_input)
confidence = np.max(proba, axis=1)
y_pred = clf.predict(X_input)

df_input["Predicted State"] = pd.Series(y_pred).map({0: "run", 1: "tumble"})
df_input["Prediction Confidence"] = confidence

# === Step 4: Save the predictions ===
df_input.to_csv("predictions_with_confidence.csv", index=False)

# === Step 5: Plot confidence histogram ===
plt.figure(figsize=(8, 5))
plt.hist(confidence, bins=10, color="steelblue", edgecolor="black")
plt.xlabel("Prediction Confidence")
plt.ylabel("Number of Samples")
plt.title("Distribution of Prediction Confidence")
plt.tight_layout()
plt.savefig("confidence_histogram.png")
plt.show()

# === Step 6: Plot confidence by predicted class ===
plt.figure(figsize=(8, 5))
sns.histplot(data=df_input, x="Prediction Confidence", hue="Predicted State", bins=10,
             multiple="stack", palette="Set1")
plt.xlabel("Prediction Confidence")
plt.ylabel("Number of Samples")
plt.title("Prediction Confidence by Predicted Class")
plt.tight_layout()
plt.savefig("confidence_by_class.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA

# -------------------------------------------------
# 1) Load full training data
# -------------------------------------------------
df  = pd.read_csv("synthetic_bacterial_chemotaxis_data.csv")
X   = df.drop(columns=["state"])
y   = df["state"].map({"run": 0, "tumble": 1})         # 0 = run, 1 = tumble

# -------------------------------------------------
# 2) Scale features & train KNN on the entire set
# -------------------------------------------------
scaler      = StandardScaler()
X_scaled    = scaler.fit_transform(X)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_scaled, y)

# -------------------------------------------------
# 3) Predict & find mis-classified points
# -------------------------------------------------
y_hat       = knn.predict(X_scaled)
mis_mask    = y_hat != y

# -------------------------------------------------
# 4) PCA projection for 2-D plot
# -------------------------------------------------
pca   = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plot_df = pd.DataFrame({
    "PC1" : X_pca[:, 0],
    "PC2" : X_pca[:, 1],
    "Pred": pd.Series(y_hat).map({0: "run", 1: "tumble"}),
    "Mis" : mis_mask
})

palette = {"run": "#d62728",   # red
           "tumble": "#1f77b4"} # blue

# -------------------------------------------------
# 5) Plot
# -------------------------------------------------
plt.figure(figsize=(10, 7))
sns.scatterplot(data=plot_df,
                x="PC1", y="PC2",
                hue="Pred", palette=palette,
                marker="o", s=75, edgecolor="none")

# Overlay misclassified points (black ring, no legend label)
err = plot_df[plot_df["Mis"]]
plt.scatter(err["PC1"], err["PC2"],
            facecolors="none", edgecolors="black",
            linewidths=1.6, s=140, label="_nolegend_")

plt.title("KNN Classification on Training Data (PCA projection)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Predicted class", bbox_to_anchor=(1.04, 1), loc="upper left")
plt.tight_layout()
plt.savefig("knn_training_cleaned_pca.png")
plt.show()
```

```python
# -------------------------------------------
# 1.  TRAIN
# -------------------------------------------
train_df = pd.read_csv("synthetic_bacterial_chemotaxis_data.csv")
feature_cols = train_df.drop(columns=["state"]).columns      # <-- save names

X_train = train_df[feature_cols]
y_train = train_df["state"].map({"run": 0, "tumble": 1})

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# -------------------------------------------
# 2.  PREDICT on new file
# -------------------------------------------
input_df  = pd.read_csv("new_bacterial_data_50.csv")
X_input   = input_df[feature_cols]            # use saved column list
X_input_s = scaler.transform(X_input)

y_pred = knn.predict(X_input_s)
proba  = knn.predict_proba(X_input_s).max(axis=1)

input_df["Predicted state"]       = pd.Series(y_pred).map({0: "run", 1: "tumble"})
input_df["Prediction confidence"] = proba
input_df.to_csv("knn_predictions_on_new_data.csv", index=False)

# -------------------------------------------
# 3.  PCA plot
# -------------------------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_input_s)

plot_df = pd.DataFrame({
    "PC1": X_pca[:, 0],
    "PC2": X_pca[:, 1],
    "Pred": input_df["Predicted state"]
})

palette = {"run": "#1f77b4", "tumble": "#d62728"}

plt.figure(figsize=(10, 7))
sns.scatterplot(data=plot_df, x="PC1", y="PC2",
                hue="Pred", palette=palette,
                s=80, edgecolor="black")
plt.title("KNN Prediction on New Input Data (PCA projection)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Predicted state", bbox_to_anchor=(1.04, 1), loc="upper left")
plt.tight_layout()
plt.savefig("input_data_classification_knn.png")
plt.show()
```
